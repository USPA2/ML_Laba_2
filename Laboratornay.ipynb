```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
```

```python
print("\nDataset Description:")
print(data.describe(include='all'))
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['y'], bins=20, kde=False, color='blue')
plt.title('Distribution of Target Variable')
plt.xlabel('Target Variable')
plt.ylabel('Frequency')
plt.show()
```

```python
print("\nBasic Statistics of Variables:")
print(data.describe())
```

```python
print("\nDataset Information:")
print(data.info())
```

```python
print("\nNumber of Missing Values:")
print(data.isnull().sum())
```

```python
features = ['Age', 'Perseverance', 'SelfConfidence', 'GoodPhysicalHealth', 'MentalDisorder']
for feature in features:
    plt.figure(figsize=(10,6))
    sns.histplot(data[feature], bins=20, kde=False, color='green')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
selected_features = data.columns[3:8].tolist()
for feature in selected_features:
    plt.figure(figsize=(10, 6))
    sns.pointplot(x='y', y=feature, data=data, dodge=True)
    plt.title(f'{feature} vs Target Variable')
    plt.xlabel('Target Variable')
    plt.ylabel(feature)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data[['Perseverance', 'SelfConfidence']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Perseverance and SelfConfidence')
plt.show()
```
